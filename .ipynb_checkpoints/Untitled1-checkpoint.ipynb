```python
import seaborn as sns
import matplotlib.pyplot as plt
#load the file
data = sns.load_dataset("iris")
#print the head of data using .head()
print(data.head())
#use of info
data.info()
#use of describe
data.describe()
# creating scatter plot for relationships between features.
plt.scatter(data.sepal_length,data.sepal_width,color='black',marker='o',label='iris Data')
plt.grid()
plt.title("Comparision Chart")
plt.show()

#histograms to show value distributions.
plt.hist(data["sepal_length"], bins=5, color='purple', edgecolor='black')
plt.xlabel('sepal_length')
plt.ylabel('sepal_width')
plt.title('value distributions')
plt.show()

#box plots to identify outliers.
sns.boxplot(x=data["sepal_length"])
plt.title("Box Plot (Outliers in Sepal Length)")
plt.show()

```
